```python
import matplotlib.pyplot as plt

# Labels
animals = ["Bears", "Dolphins", "Whales"]

# Data for each year
data = {
    "2017": [8, 150, 80],
    "2018": [54, 77, 54],
    "2019": [93, 32, 100],
    "2020": [116, 11, 76],
    "2021": [137, 6, 93],
    "2022": [184, 1, 72]
}

# Loop to create pie chart for each year
for year, values in data.items():
    plt.figure()  # new window for each chart
    plt.pie(values, labels=animals, autopct='%1.1f%%')
    plt.title(f"Animal Population in {year}")
    plt.show()
```

b)

```python
import matplotlib.pyplot as plt

animals = ["Bears", "Dolphins", "Whales"]

data = {
    "2017": [8, 150, 80],
    "2018": [54, 77, 54],
    "2019": [93, 32, 100],
    "2020": [116, 11, 76],
    "2021": [137, 6, 93],
    "2022": [184, 1, 72]
}

# Create subplots (2 rows, 3 columns)
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes = axes.flatten()  # make it easier to loop

for i, (year, values) in enumerate(data.items()):
    axes[i].pie(values, labels=animals, autopct='%1.1f%%')
    axes[i].set_title(year)

plt.tight_layout()
plt.show()
```

f) Test out the following pie chart additional arguments / settings (“...” is for the normal
code you have already)
A myexplode = [0.3, 0, 0, 0]
plt.pie(... , explode = myexplode)
B myexplode = [0.3, 0, 0, 0]
plt.pie(... , explode = myexplode,shadow = True)
C plt.legend(title = "Car Colours:")
D mycolours = ["Magenta", "Green", "Cyan", "Red"]
plt.pie(... , colors = mycolours)

```python
import matplotlib.pyplot as plt

cars = ["Red", "Silver", "Blue", "White", "Green"]
values = [123, 230, 78, 193, 12]

# A - explode
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(values, labels=cars, explode=myexplode)
plt.title("A - Explode")
plt.show()

# B - explode + shadow
plt.pie(values, labels=cars, explode=myexplode, shadow=True)
plt.title("B - Explode + Shadow")
plt.show()

# C - legend
plt.pie(values, labels=cars)
plt.legend(title="Car Colours:")
plt.title("C - Legend")
plt.show()

# D - custom colors
mycolours = ["magenta", "green", "cyan", "red", "blue"]
plt.pie(values, labels=cars, colors=mycolours)
plt.title("D - Custom Colours")
plt.show()
```
