```python
def fib(n):
    a = 0
    b = 1
    count = 0

    while count < n:
        print(a)
        c = a + b
        a = b
        b = c
        count = count + 1

# Example
n = int(input("Enter number of terms: "))
fib(n)
```

```python
def myPow(x, n):
    result = 1
    count = 0

    while count < n:
        result = result * x
        count = count + 1

    return result

# Example
print(myPow(2,3))
```

```python
def SumDigits(n):
    total = 0

    while n > 0:
        digit = n % 10
        total = total + digit
        n = n // 10

    print(total)
SumDigits(234)
```

```python
def squaredSum(arr):
    total = 0
    i = 0

    while i < len(arr):
        total = total + arr[i] * arr[i]
        i = i + 1

    return total
print(squaredSum([2,3,4]))
```

```python
def sumDigits(num):
    total = 0

    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10

    return total
print(sumDigits(678))
```
