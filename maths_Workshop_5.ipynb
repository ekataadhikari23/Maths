np.dot():

Works for both scalar multiplication and matrix multiplication.
Can behave differently for 1D, 2D, or higher arrays.

np.matmul():

Used only for matrix multiplication.
Does NOT work with scalars.

Result of np.dot(5, arr1)
It multiplies every element of arr1 by 5.

```python
import numpy as np
arr1 = np.array([[1,2,3],
                 [2,3,4],
                 [2,3,4]])

arr2 = np.array([[21,4],
                 [2,1],
                 [2,3]])
arr3 = np.dot(arr1, arr2)
transpose_arr3 = np.transpose(arr3)
print("arr3:")
print(arr3)
print("Transpose of arr3:")
print(transpose_arr3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Vectors
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors from origin
ax.quiver(0,0,0, U[0],U[1],U[2])
ax.quiver(0,0,0, V[0],V[1],V[2])
ax.quiver(0,0,0, W[0],W[1],W[2])

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
import numpy as np
def direction_cosines(v):
    mag = np.sqrt(np.sum(v**2))
    return v / mag
O = np.array([1.5, 1.5, 2.0])
print("Direction cosines:", direction_cosines(O))
```
